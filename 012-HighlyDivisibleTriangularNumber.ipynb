## Highly divisible triangular number

The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

Let us list the factors of the first seven triangle numbers:

     1: 1
     3: 1,3
     6: 1,2,3,6
    10: 1,2,5,10
    15: 1,3,5,15
    21: 1,3,7,21
    28: 1,2,4,7,14,28

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

https://www.dcode.fr/divisors-list-number


```python
def isPrime(n):
    result = True
    for i in range(2,n):
        if(n%i == 0):
            result = False
            break
    return result
```

```python
def get_divisors(n):
    res = [1]
    for i in range(2, int(n/2)+1):
        if(n%i == 0):
            res.append(i)
    return res
```

```python
for i in range(7524,20000):
    triangN = 0
    for j in range(1,i+1):
        triangN += j
    if isPrime(triangN) == False:
      size = len(get_divisors(triangN))
      if size > 250:
        print(i, ": triangular number=", triangN, " size:", size)
      if(size > 500):
        print("FOUND: ", i, ": triangular number=", triangN)
        print(get_divisors(triangN))
        exit()
    
```
